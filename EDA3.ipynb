```python
# plotting a box plot for Purchased Amount by Frequency of Purchases
plt.figure(figsize= (10, 10))
df.boxplot(column ='Purchase Amount (USD)',by = "Frequency of Purchases")
plt.title("")
plt.show()
```

ðŸ“¦ Boxplot: Purchase Amount by Frequency of Purchases

â€¢ This boxplot illustrates how purchase amounts (USD) vary across different purchase frequencies in the dataset.

ðŸ”§ Plot Details:

â€¢ Plot Type: Boxplot (df.boxplot(column='Purchase Amount (USD)', by='Frequency of Purchases'))

â€¢ Figure Size: 10 Ã— 10 inches

â€¢ X-axis: Frequency of Purchases (e.g., Weekly, Monthly, Quarterly)

â€¢ Y-axis: Purchase Amount (USD)

â€¢ Title: "Boxplot grouped by Frequency of Purchases"

ðŸ“ˆ Insights:

â€¢ Each box shows the median (green line), interquartile range (box), and outliers (whiskers).

â€¢ Purchase behavior varies across frequencies:

   Â° Weekly and Monthly shoppers tend to have higher variability in spending.

   Â° Annual and Quarterly shoppers show tighter ranges, suggesting more consistent spending patterns.

â€¢ This visualization helps identify how often customers shop and how much they typically spend, which is useful for segmentation and targeting strategies.

```python
# Count of Payment Method
# Count the values
counts = df['Payment Method'].value_counts()

# Generate distinct colors
num_bars = len(counts)
colors = plt.cm.tab20(np.linspace(0, 1, num_bars))

# Create a plot and save the axes object
ax = counts.plot(kind='bar', edgecolor='black', color=colors)

# Add title and labels (optional)
plt.ylabel('Count')
plt.title('---')

# Annotate each bar with its value
for i, value in enumerate(counts):
    ax.text(i, value + 1, str(value), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

ðŸ“¦ Boxplot: Purchase Amount by Frequency of Purchases

â€¢ This boxplot illustrates how purchase amounts (in USD) vary across different purchase frequencies in the dataset.

ðŸ”§ Plot Details:

â€¢ Plot Type: Boxplot (df.boxplot(column='Purchase Amount (USD)', by='Frequency of Purchases'))

â€¢ Figure Size: 10 Ã— 10 inches

â€¢ X-axis: Frequency of Purchases (e.g., Weekly, Monthly, Quarterly)

â€¢ Y-axis: Purchase Amount (USD)

â€¢ Title: "Boxplot grouped by Frequency of Purchases"

ðŸ“ˆ Insights:

â€¢ Each box shows the median (green line), interquartile range (box), and outliers (whiskers).

â€¢ Spending patterns vary by frequency:

  Â° Weekly and Monthly shoppers show wider ranges and higher variability in purchase amounts.

  Â° Annual and Quarterly shoppers tend to have more consistent spending, with narrower boxes and fewer outliers.
